```python
from esel3d import plot3d
import numpy as np
```

```python
#setup domain parameters
xinitial = -2 
xfinal = 2 

yinitial = -2 
yfinal = 2 

#the finer the resolution, the smoother the graph but cost more in computation.
resolution = .1
```

```python
#create the function to plot.
#in this case, this function will produce the "saddle".
def func(x, y):
  return np.power(x, 2) - np.power(y, 2)
def func2(x, y):
  return -np.power(x, 2)
def func3(x, y):
  return x
```

```python
#plot3d() instance must be in the same cell as plot()

plt3d = plot3d()
plt3d.surface(func, [xinitial, xfinal], [yinitial, yfinal], resolution)
plt3d.surface(func2, [0, xfinal], [0, yfinal], resolution)
plt3d.surface(func3, [xinitial, xfinal], [yinitial, yfinal], resolution)
plt3d.plot()

```

```python
del plt3d
```

```python
def func4(x,y):
  return np.sqrt(4-np.power(x,2))
def func5(x,y):
  return -np.sqrt(4-np.power(x,2))
def func6(x,y):
  return np.sqrt(4-np.power(y,2))
def func7(x,y):
  return -np.sqrt(4-np.power(y,2))
xfunc = [func4, func5]
yfunc = [func6, func7]
  
```

```python
plt = plot3d()
plt.surface(xfunc, [xinitial, xfinal], [yinitial, yfinal], resolution)
plt.surface(yfunc, [xinitial,xfinal], [yinitial, yfinal], resolution)
plt.plot()
```
